# Prop 67 Analysis

An analysis of Proposition 67, which is a California ballot proposition on the November 8, 2016 ballot. A "Yes" vote was to approve, and a "No" vote to reject, a statute that prohibits grocery and other stores from providing customers singleâ€“use plastic or paper carryout bags but permits sale of recycled paper bags and reusable bags.

Proposition 67 passed with 53% of the vote. With its passage, Senate Bill 270, which bans plastic shopping bags, was upheld and went into effect. A related measure, Proposition 65, failed at the polls.

### Introduced a new column "contributor_fullname"

```python
merged["contributor_fullname"] = merged["contributor_firstname"].map(str) + " " + merged["contributor_lastname"].map(str)
```

```python
merged["contributor_fullname"] = merged["contributor_fullname"].str.lstrip('nan')
```
